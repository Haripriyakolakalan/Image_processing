```python
limit=40
tim_images=[None]*limit
j=0
for i in tim:
  if(j<limit):
    tim_images[j]=imread("/content/drive/MyDrive/cv2/Tim Cook/"+i)
    j+=1
  else:
    break
```

```python
for i in range(len_of_images_sundar):
  sundar_gray[i]=np.ndarray.flatten(sundar_gray[i].reshape(flatten_size_sundar,1))
```

```python
for i in range(len_of_images_tim):
  tim_gray[i]=np.ndarray.flatten(tim_gray[i].reshape(flatten_size_tim,1))
```

```python
sundar_gray=np.dstack(sundar_gray)
```

```python
sundar_gray=sundar_gray.reshape(len_of_images_sundar,flatten_size_sundar)
sundar_gray.shape
```

```python
tim_gray=tim_gray.reshape(len_of_images_tim,flatten_size_tim)
tim_gray.shape
```

```python
sundar_data=pd.DataFrame(sundar_gray)
sundar_data
```

```python
tim_data=pd.DataFrame(tim_gray)
tim_data
```

```python
sundar_data["label"]="Sundar Pichai"
sundar_data
```

```python
tim_data["label"]="Tim Cook"
tim_data
```

```python
total=pd.concat([sundar_data,tim_data])
total
```

```python
from sklearn.utils import shuffle
```

```python
data_indexed=shuffle(total).reset_index()
data_indexed
```

```python
data=data_indexed.drop(['index'],axis=1)
data
```

```python
x=data.values[:,:-1]
y=data.values[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
```

```python
clf=svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```
